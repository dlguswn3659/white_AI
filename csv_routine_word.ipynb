```python
from konlpy.corpus import kobill
from konlpy.tag import Okt
import numpy as np
from matplotlib import pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import pandas as pd

file_path_name = ['/content/find_mm.txt', '/content/find_sm2.txt', '/content/find_mmsm.txt']
graph_title = ['노이즈 많은 매매', '노이즈 많은 세무', '노이즈 많은 매매-세무']
data_size = []

mmy_smy = []  #매매로도 o, 세무로도 o
mmn_smy = []  #매매는 x, 세무로는 o
smy = []      #매매로는 ?, 암튼 세무는 o

for data_num in range(0, len(file_path_name)):
  t = Okt()
  doc_ko = open(file_path_name[data_num], encoding='utf-8')
  data = doc_ko.readlines()

  csv_word = []
  tokens_ko = []
  # print(len(data))
  data_size.append(len(data))

  for line in data:
    csv_word_tmp = t.nouns(line)
    csv_word.append(csv_word_tmp)

  for i in range(0, len(csv_word)):
    for j in range(0, len(csv_word[i])):
      tokens_ko.append(csv_word[i][j])

  # print(tokens_ko)

  ko = nltk.Text(tokens_ko, name = graph_title[data_num])
  # print(len(ko.tokens))
  # print(len(set(ko.tokens)))
  ko.vocab()

  ko = nltk.Text(ko, name=graph_title[data_num])
  plt.figure(figsize=(12,6))
  ko.plot(50)
  # plt.ylabel('빈도수', fontproperties=fontprop)
  # plt.savefig('mmsm' + str(data_num) + '.png')
  plt.show()

  data2 = ko.vocab().most_common(50)
  # print(data2)
  # print(type(data2))
  # print(data2[3][1])
  if data_num == 0:
    mmy_smy = data2
  elif data_num == 1:
    mmn_smy = data2
  elif data_num == 2:
    smy = data2
  df = pd.DataFrame(data2)
  print(df)
  wordcloud = WordCloud(font_path = '/content/malgun.ttf', relative_scaling = 0.2, background_color='white',).generate_from_frequencies(dict(data2))
  plt.figure(figsize=(12,8))
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.show()
```

```python
print(mmy_smy)
print(mmn_smy)
print(smy)

print(len(mmy_smy))
```

```python
one_two_score = []
only_first_case = []
only_second_case = []

print(data_size)

for i in range(0, len(mmy_smy)):
  is_exist = 0
  for j in range(0, len(mmn_smy)):
    if mmy_smy[i][0] == mmn_smy[j][0]:
      word_tmp = []
      word_tmp.append(mmy_smy[i][0])
      mmy_smy_point = mmy_smy[i][1] / data_size[0]
      mmn_smy_point = mmn_smy[j][1] / data_size[1]
      total_point = mmn_smy_point / mmy_smy_point
      word_tmp.append(total_point)
      one_two_score.append(word_tmp)
      is_exist = 1
  if is_exist == 0:
    only_first_case.append(mmy_smy[i][0])

for i in range(0, len(mmn_smy)):
  is_exist = 0
  for j in range(0, len(mmy_smy)):
    if mmn_smy[i][0] == mmy_smy[j][0]:
      is_exist = 1
  if is_exist == 0:
    only_second_case.append(mmn_smy[i][0])

print(one_two_score)
print(type(one_two_score))

df = pd.DataFrame(one_two_score)

print(only_first_case)
print(only_second_case)
print(mmy_smy)
print(mmn_smy)

# sorted(one_two_score[][1], reverse=True)
# print(df)
# df.sort(axis=1, ascending=False) 
df.sort_values(by=[1], axis=0, ascending=False)
```

```python
from konlpy.corpus import kobill
from konlpy.tag import Okt
import numpy as np
from matplotlib import pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import pandas as pd

file_path_name = ['/content/find_br.txt', '/content/find_hj.txt', '/content/find_only_br.txt', '/content/find_only_hj.txt']
graph_title = ['노이즈 많은 법률', '노이즈 많은 행정', '노이즈 많은 법률(행정 미포함)', '노이즈 많은 행정(법률 미포함)']
data_size = []

bry = []      #법률 o
hjy = []      #행정 o
bry_hjn = []  #법률 o, 행정 x
brn_hjy = []  #법률 x, 행정 o

for data_num in range(0, len(file_path_name)):
  t = Okt()
  doc_ko = open(file_path_name[data_num], encoding='utf-8')
  data = doc_ko.readlines()

  csv_word = []
  tokens_ko = []
  # print(len(data))
  data_size.append(len(data))

  for line in data:
    csv_word_tmp = t.nouns(line)
    csv_word.append(csv_word_tmp)

  for i in range(0, len(csv_word)):
    for j in range(0, len(csv_word[i])):
      tokens_ko.append(csv_word[i][j])

  # print(tokens_ko)

  ko = nltk.Text(tokens_ko, name = graph_title[data_num])
  # print(len(ko.tokens))
  # print(len(set(ko.tokens)))
  ko.vocab()

  ko = nltk.Text(ko, name=graph_title[data_num])
  plt.figure(figsize=(12,6))
  ko.plot(50)
  # plt.ylabel('빈도수', fontproperties=fontprop)
  # plt.savefig('mmsm' + str(data_num) + '.png')
  plt.show()

  data2 = ko.vocab().most_common(50)
  # print(data2)
  # print(type(data2))
  # print(data2[3][1])
  if data_num == 0:
    bry = data2
  elif data_num == 1:
    hjy = data2
  elif data_num == 2:
    bry_hjn = data2
  elif data_num == 3:
    brn_hjy = data2
    
  df = pd.DataFrame(data2)
  print(df)
  wordcloud = WordCloud(font_path = '/content/malgun.ttf', relative_scaling = 0.2, background_color='white',).generate_from_frequencies(dict(data2))
  plt.figure(figsize=(12,8))
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.show()
```

```python
one_two_score = []
only_first_case = []
only_second_case = []

print(data_size)

for i in range(0, len(bry)):
  is_exist = 0
  for j in range(0, len(hjy)):
    if bry[i][0] == hjy[j][0]:
      word_tmp = []
      word_tmp.append(bry[i][0])
      bry_point = bry[i][1] / data_size[0]
      hjy_point = hjy[j][1] / data_size[1]
      total_point = hjy_point / bry_point
      word_tmp.append(total_point)
      one_two_score.append(word_tmp)
      is_exist = 1
  if is_exist == 0:
    only_first_case.append(bry[i][0])

for i in range(0, len(hjy)):
  is_exist = 0
  for j in range(0, len(bry)):
    if hjy[i][0] == bry[j][0]:
      is_exist = 1
  if is_exist == 0:
    only_second_case.append(hjy[i][0])

print(one_two_score)
print(type(one_two_score))

df = pd.DataFrame(one_two_score)

print(only_first_case)
print(only_second_case)
print(bry)
print(hjy)

# sorted(one_two_score[][1], reverse=True)
# print(df)
# df.sort(axis=1, ascending=False) 
df.sort_values(by=[1], axis=0, ascending=False)
```
